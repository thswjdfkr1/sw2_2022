```python
class Calculator:
  def __init__(self):
    self.value = 0

  def add(self, val):
    self.value += val
```

```python
class UpgradeCalculator():
  def minus(self, val2):
    self.value -= val2
```

```python
class maxLimitCalculator(Calculator):
  def add(self,val):
    self.value += val
    if self.value > 100:
      self.value = 100
```

```python
a= maxLimitCalculator()
a.add(60)
a.add(70)

print(a.value)
```
