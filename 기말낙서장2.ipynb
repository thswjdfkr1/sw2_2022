```python
class country2:
  cn_code = []
  cn_pop = {}

  def __init__(a, u_code, u_pop, u_gdp):
    a.code = u_code
    a.pop=u_pop
    a.gpd= u_gdp
    country2.cn_code.append(a.code)
    country2.cn_pop[a.code]=a.pop
```
